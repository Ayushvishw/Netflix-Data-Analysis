### Importing Matplotlib library

```python
from matplotlib import style
style.use("ggplot")
my_colors = ['maroon','darkred','black', 'darkred','darkred','maroon','black', 'darkgrey', 'grey']
plt.bar(rls_yr.index, rls_yr, color= my_colors)
plt.title(" HERO YEARS WITH MOST OF RELEASES", fontsize = 20)
plt.xlabel("Year Of Release", fontsize = 11)
plt.ylabel("No. Of Movies & Shows", fontsize = 11)
```

### Q4 How many movies and tv shows are in the datassets ? show with a bar graph ? 

### .groupby()

```python
types = data.groupby("type").type.count()
types
```

```python
from matplotlib import style 
style.use("ggplot")
my_colors = ['maroon','black']
types.plot(kind = "bar", color  = my_colors, width = 0.5, edgecolor = "k", linewidth = 4)
plt.title("COUNT OF TV SHOWS & MOVIES", fontsize = 20)
plt.xlabel("TYPE", fontsize = 14)
plt.ylabel("COUNT", fontsize = 14)
plt.grid('k-', linewidth = 2)                                      # Using matplotlib       
```

### Q5) Find the most number of count of duration from the datasets ? Show with a pie chart ? 

```python
data.head()
```

```python
durations = data.groupby("duration").duration.count().sort_values(ascending=False).head()
durations
```

```python
# Creating a variable and storing the colors fro pie chart
my_colors = ['maroon','darkred','black', 'darkgrey', "grey"]
# Storing the labels in the varibles 
my_labels = '1 Season', '2 Seasons' ,'3 Seasons', '90 min', '97 min' 
# Providing the explode size for pie chart
my_explode = (0, 0.1, 0.1, 0.2, 0.2)
plt.pie(durations ,autopct='%1.1f%%', startangle=20, labels = my_labels, shadow = True, colors=my_colors, explode=my_explode)
plt.title('TOP 5 DURATIONS OF MOVIES AND SHOWS')
plt.axis('equal')
plt.legend(my_labels)
plt.show()  #using pie  
```

### Q7)  Find top 5 countries with most movies and tv shows ? Show with pie chart ? 

```python
county = data.groupby("country").country.count().sort_values(ascending = False).head()
county
```

```python
# Creating a variable and storing the colors fro pie chart
my_colors = ['maroon','darkred','black', 'darkgrey', "grey"]
# Storing the labels in the varibles 
my_labels = 'United States', 'India' ,'3 United Kingdom', 'Japan', 'South Korea' 
# Providing the explode size for pie chart
my_explode = (0, 0.2, 0.2, 0.3, 0.6)
plt.pie(county ,autopct='%1.1f%%', startangle=100, labels = my_labels, shadow = True, colors=my_colors, explode=my_explode)
plt.title('TOP 5 COUNTRIES WITH MOSTH OF RELEASES')
plt.axis('equal')
plt.show()  #using pie  
```
